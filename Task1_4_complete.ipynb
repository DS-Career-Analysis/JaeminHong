```python
import pandas as pd
import numpy as np
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
browser = webdriver.Chrome()
browser.get('https://www.wanted.co.kr')
```

```python
elem = browser.find_element(By.CLASS_NAME, 'searchButton')
elem.click()
elem = browser.find_element(By.CLASS_NAME, 'SearchBar_SearchBar_container__kpoXq > form > input')
elem.send_keys('데이터')
elem.send_keys(Keys.ENTER)
```

```python
last_height = browser.execute_script('return document.body.scrollHeight')  
while True:
    # 스크롤 무빙                                                    
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight);')

    # 페이지 로드 대기
    time.sleep(1)                                               
    browser.execute_script('window.scrollTo(0, document.body.scrollHeight-50);')  
    time.sleep(1)

    # 새로운 스크롤 동작 높이와 이전 스크롤과의 비교  
    new_height = browser.execute_script('return document.body.scrollHeight')

    if new_height == last_height:                                               
        break

    last_height = new_height

# 참고: https://moondol-ai.tistory.com/107
```

## **직무, 회사, 링크**

```python
Pos = [] # 직무
Com = [] # 회사
Lin = [] # 링크

i = 1
while True:
    try:
        pos = browser.find_element(By.XPATH, f'//*[@id="__next"]/div[4]/div/div[2]/div[3]/div/div/div[2]/ul/li[{i}]/div/a/div/div[1]').text
        com = browser.find_element(By.XPATH, f'//*[@id="__next"]/div[4]/div/div[2]/div[3]/div/div/div[2]/ul/li[{i}]/div/a/div/div[2]').text
        lin = browser.find_element(By.XPATH, f'//*[@id="__next"]/div[4]/div/div[2]/div[3]/div/div/div[2]/ul/li[{i}]/div/a').get_attribute('href')
        
        i += 1
        
        Pos.append(pos)
        Com.append(com)
        Lin.append(lin)
    except Exception as e:
        break

a = Pos[0]
b = len(Pos)
c = Com[0]
d = len(Com)
e = Lin[0]
f = len(Lin)

print('{}, {}\n{}, {}\n{}, {}'.format(a, b, c, d, e, f))
```

## **위치, 자격요건, 우대사항**

```python
Loc = [] # 위치
Req = [] # 자격요건
Pre = [] # 우대사항

for link in Lin:
    browser.get(link)
    time.sleep(1)
    try:
        loc = browser.find_element(By.XPATH, '//*[@id="__next"]/div[3]/div[1]/div[1]/div/section[2]/div[1]/span').text
        req = browser.find_element(By.XPATH, '//*[@id="__next"]/div[3]/div[1]/div[1]/div/div[2]/section[1]/p[3]/span').text
        pre = browser.find_element(By.XPATH, '//*[@id="__next"]/div[3]/div[1]/div[1]/div/div[2]/section[1]/p[4]/span').text
        
        Loc.append(loc)
        Req.append(req)
        Pre.append(pre)
    except Exception as e:
        break

a = Loc[0]
b = len(Loc)
c = Req[0]
d = len(Req)
e = Pre[0]
f = len(Pre)

print('{}, {}\n{}, {}'.format(a, b, c, d, e, f))
```

```python
wanted_DATA_df = pd.DataFrame({
    'Position' : Pos,     # 직무
    'Company' : Com,      # 회사
    'Requirement' : Req,  # 자격요건
    'Preference' : Pre,   # 우대사항
    'Location' : Loc,     # 위치
    'Link' : Lin          # 링크
})

wanted_DATA_df
```
